```python
import pandas as pd
```

```python
datasets = ['lacafe', 'fortuna', 'offcombr', 'hatebr']

for dataset in datasets:
    # Constantes Lacafe
    if dataset == 'lacafe':
        PATH = './Datasets/Lacafe' # Caminho para a pasta onde está o arquivo do dataset
        ARCHIVE_NAME = 'df_dataset' # Nome do arquivo do dataset sem extensao
        EXTENSION = 'csv' # Extensao do arquivo

        TEXT_COLUMN = 'txt' # Nome da coluna no arquivo que representa o texto
        CLASS_COLUMN = 'has_anger' # Nome da coluna no arquivo que representa a classe ou rotulo

    # Constantes Fortuna
    if dataset == 'fortuna':
        PATH = './Datasets/FortunaUpdated' # Caminho para a pasta onde está o arquivo do dataset
        ARCHIVE_NAME = '2019-05-28_portuguese_hate_speech_binary_classification' # Nome do arquivo do dataset sem extensao
        EXTENSION = 'csv' # Extensao do arquivo

        TEXT_COLUMN = 'text' # Nome da coluna no arquivo que representa o texto
        CLASS_COLUMN = 'hatespeech_comb' # Nome da coluna no arquivo que representa a classe ou rotulo

    # Constantes ToLD-BR
#     if dataset == 'toldbr':
#         PATH = './Datasets/ToldBR' # Caminho para a pasta onde está o arquivo do dataset
#         ARCHIVE_NAME = 'ToLD-BR-Treated' # Nome do arquivo do dataset sem extensao
#         EXTENSION = 'csv' # Extensao do arquivo

#         TEXT_COLUMN = 'text' # Nome da coluna no arquivo que representa o texto
#         CLASS_COLUMN = 'hatespeech' # Nome da coluna no arquivo que representa a classe ou rotulo

    # Constantes OffComBr
    if dataset == 'offcombr':
        PATH = './Datasets/OffComBR' # Caminho para a pasta onde está o arquivo do dataset
        ARCHIVE_NAME = 'OffComBR2' # Nome do arquivo do dataset sem extensao
        EXTENSION = 'csv' # Extensao do arquivo

        TEXT_COLUMN = 'text' # Nome da coluna no arquivo que representa o texto
        CLASS_COLUMN = 'offensive' # Nome da coluna no arquivo que representa a classe ou rotulo
        
        
    # Constantes HateBR
    if dataset == 'hatebr':
        PATH = './Datasets/HateBR' # Caminho para a pasta onde está o arquivo do dataset
        ARCHIVE_NAME = 'HateBR' # Nome do arquivo do dataset sem extensao
        EXTENSION = 'csv' # Extensao do arquivo

        TEXT_COLUMN = 'instagram_comments' # Nome da coluna no arquivo que representa o texto
        CLASS_COLUMN = 'offensive_language' # Nome da coluna no arquivo que representa a classe ou rotulo
    
    data = pd.read_csv(PATH + '/' + ARCHIVE_NAME + '.' + EXTENSION)

    if dataset == 'lacafe':
        data = data[data['docno'] != 20410]

    print(dataset)
        
    # Monta o caminho para salvar o arquivo
    for i in range(3):
        train_path = f'{PATH}/{ARCHIVE_NAME}_train_{i+1}.{EXTENSION}'
        test_path = f'{PATH}/{ARCHIVE_NAME}_test_{i+1}.{EXTENSION}'
        train_data = pd.read_csv(train_path)
        test_data = pd.read_csv(test_path)
        train_positive_examples = len(train_data[train_data[CLASS_COLUMN] == 1].values)
        train_negative_examples = len(train_data[train_data[CLASS_COLUMN] == 0].values)
        train_total = len(train_data.values)
        test_positive_examples = len(test_data[test_data[CLASS_COLUMN] == 1].values)
        test_negative_examples = len(test_data[test_data[CLASS_COLUMN] == 0].values)
        test_total = len(test_data.values)
        print(str(i+1) + ' Treino Positivos: ' + str(train_positive_examples))
        print(str(i+1) + ' Treino Negativos: ' + str(train_negative_examples))
        print(str(i+1) + ' Treino Total: ' + str(train_total))
        print(str(i+1) + ' Teste Positivos: ' + str(test_positive_examples))
        print(str(i+1) + ' Teste Negativos: ' + str(test_negative_examples))
        print(str(i+1) + ' Teste Total: ' + str(test_total))
        
    # Descobre quandos exemplos de cada classe existem
    total_positive_examples = len(data[data[CLASS_COLUMN] == 1].values)
    total_negative_examples = len(data[data[CLASS_COLUMN] == 0].values)
    total = len(data.values)

    print('Positivos: ' + str(total_positive_examples))
    print('Negativos: ' + str(total_negative_examples))
    print('Total: ' + str(total))
    print()
```
